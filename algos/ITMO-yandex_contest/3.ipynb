```python
import pandas as pd
from scipy.stats import binom
import math
```

```python
df = pd.read_csv('./sample_final.csv')
df.shape
```

```python
thursday = df[df['День недели'] == 'Четверг']
thursday
```

```python
thursday = thursday[thursday['Видеозапись'] == 'Видосы про собачек']
count = len(thursday)
print(count)
thursday
```

```python
prob_8_watch = binom.pmf(8, 100, count / 100)
round(prob_8_watch, 4)
```

```python
prob_more_than_8_watch = 1 - binom.cdf(8, 100, count / 100)
round(prob_more_than_8_watch, 4)
```
