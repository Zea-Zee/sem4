```python
from operator import itemgetter
from mmaction.apis import init_recognizer, inference_recognizer


def pred(video_path, top_n=2):
    print(f"Video: {video_path}")
    config_file = '/home/zea/Desktop/sem4/PHC/labs/lab4(mmaction)/tsn_imagenet-pretrained-r50_8xb32-1x1x3-100e_kinetics400-rgb.py'
    checkpoint_file = '/home/zea/Desktop/sem4/PHC/labs/lab4(mmaction)/mmaction2/work_dirs/tsn_imagenet-pretrained-r50_8xb32-1x1x3-100e_kinetics400-rgb/epoch_100.pth'
    video_file = video_path
    label_file = '/home/zea/Desktop/sem4/PHC/labs/lab4(mmaction)/data/kinetics400_tiny/label_map.txt'
    model = init_recognizer(config_file, checkpoint_file, device='cpu')  # or device='cuda:0'
    pred_result = inference_recognizer(model, video_file)

    pred_scores = pred_result.pred_score.tolist()
    score_tuples = tuple(zip(range(len(pred_scores)), pred_scores))
    score_sorted = sorted(score_tuples, key=itemgetter(1), reverse=True)
    top2_label = score_sorted[:2]

    labels = open(label_file).readlines()
    labels = [x.strip() for x in labels]
    results = [(labels[k[0]], k[1]) for k in top2_label]

    print(f'The top-{top_n} labels with corresponding scores are:')
    for result in results:
        print(f'{result[0]}: ', result[1])


pred('/home/zea/Desktop/sem4/PHC/labs/lab4(mmaction)/data/kinetics400_tiny/test/communist_explaining_why_its_not_shit.mp4')
pred('/home/zea/Desktop/sem4/PHC/labs/lab4(mmaction)/data/kinetics400_tiny/test/putin_and_russia.mp4')
```
